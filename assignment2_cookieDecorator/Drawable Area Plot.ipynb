```python
import numpy as np
import matplotlib.pyplot as plt

# Arm segment lengths in cm
L1 = 8.5
L2 = 10.0

# Servo base location on the platform: centered on left side (x=0, y=half platform height)
servo_x = 0
servo_y = 5

# Sample the joint angles within their limits:
# θ₁ (shoulder) from -90° to 90° -> -π/2 to π/2
# θ₂ (elbow) from -180° to 0° -> -π to 0
theta1_vals = np.linspace(-np.pi/2, np.pi/2, 300)
theta2_vals = np.linspace(-np.pi, 0, 300)

# Compute reachable (x, y) positions relative to servo base
X = []
Y = []
for theta1 in theta1_vals:
    for theta2 in theta2_vals:
        x = L1 * np.cos(theta1) + L2 * np.cos(theta1 + theta2)
        y = L1 * np.sin(theta1) + L2 * np.sin(theta1 + theta2)
        # Translate the coordinate system so that the servo base is at (0, 5)
        X.append(x + servo_x)
        Y.append(y + servo_y)

# Plot the reachable area and the platform
plt.figure(figsize=(8,6))
plt.scatter(X, Y, s=1, color='blue', alpha=0.5, label="Reachable Area")

# Draw the platform: 15 cm wide and 10 cm tall, with bottom-left at (0,0)
platform_x = [0, 15, 15, 0, 0]
platform_y = [0, 0, 10, 10, 0]
plt.plot(platform_x, platform_y, color='red', linewidth=2, label="Platform")

# Mark the servo's base location
plt.plot(servo_x, servo_y, 'ko', markersize=8, label="Servo Base")

plt.xlim(-5, 20)
plt.ylim(-5, 20)
plt.xlabel("X (cm)")
plt.ylabel("Y (cm)")
plt.title("Reachable Area of the Drawing Robot and Platform")
plt.legend()
plt.grid(True)
plt.show()
```
